```python
%matplotlib inline
import os
import subprocess

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Pegar o nome dos executaveis
executables = sorted([n for n in os.listdir("./build/") if n.startswith('vector') and n != 'vector_operations'])
executables
```

```python
def plot_graph(executable):
    output = subprocess.check_output('./build/' + executable)
    output = output.decode("utf-8")
    data = [x.split(',') for x in output.split('\n')][:][:-1] # O ultimo estava vazio, entao tirei
    df = pd.DataFrame(data, columns=["Experiment", "n", "mean", "std"], dtype=np.float64)
    groups = df.groupby("Experiment")
    fig, ax = plt.subplots()
    for name, group in groups:
        ax.plot(group['n'], group['mean'], marker='o', linestyle='-', ms=5, label=name)
    plt.title('Tempos para ' + executable)
    plt.ylabel('Tempo (ms)')
    plt.axis([-1000, 1000000+10000, -1, 35])
    plt.xlabel('Tamanho do vetor (n)')
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.show()
```

```python
for e in executables:
    plot_graph(e)
```
